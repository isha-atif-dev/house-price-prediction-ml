Above the 1.2 M price could be outliers.

# Correlation

Do two things move together?

If area increases, does price increase?

- Correlation = relationship between two variables

# Scatter Plot (Area vs Price)

This shows relationship between two variables.

```python
plt.scatter(df["area"], df["price"])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.show()
```

- Bigger houses tend to cost more.

This is called positive correlation.

# Correlation Heatmap

Which features influence price the most?

```python

df_numeric = df.select_dtypes(include=["int64","float64"])

plt.figure(figsize=(6,4))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm")

plt.title("Feature Correlation Heatmap")
plt.show()
```

price ↔ area       = 0.54

price ↔ bathrooms  = 0.52

price ↔ stories    = 0.42

price ↔ parking    = 0.38

price ↔ bedrooms   = 0.37

- Area (0.54)

This is the strongest relationship with price.

# Boxplot (Find Outliers)

- Boxplots help detect outliers.

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["price"])
plt.title("Price Boxplot")
plt.show()
```
